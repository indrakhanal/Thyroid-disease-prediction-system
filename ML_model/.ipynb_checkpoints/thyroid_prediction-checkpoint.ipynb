### Importing Necessary Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('dataset.csv', header=None)
data_frame.head()
```

### Adding Heading to the Dataset

```python
data_frame2 = pd.read_excel('new-header.xlsx', header=None)
temp = []
for i in range(0, len(data_frame2[0])):
    temp.append(data_frame2[0][i])
# print(temp)
data_frame.columns=temp
data_frame.head()
```

### 1, 2, 3 denotes the classes of thyroid into whiich class 1 denotes patient with normal condition and 2 for hyperthyroidsm and 3 for hypothyroidism 

```python
a = data_frame.shape
print(a)
data_frame['Class'].value_counts()
```

### Shape of the data

```python
hypo_thyroid = data_frame[data_frame['Class']==1][0:90]
negative_thyroid = data_frame[data_frame['Class']==2][0:150]
hyper_thyroid = data_frame[data_frame['Class']==3][0:2000]
# hypo_thyroid
x_axis = hypo_thyroid.plot(kind='scatter', x='TSH', y='T4U', color='blue',label='hyperthoride')
y_axis = negative_thyroid.plot(kind='scatter', x='TSH', y='T4U', color='red',label='hypothyroid')
z_axis = hyper_thyroid.plot(kind='scatter', x='TSH', y='T4U', color='black',label='Normal')
```
